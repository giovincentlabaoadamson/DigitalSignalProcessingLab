# Laboratory Experiment 6

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Problem 1

```python
n_samples = 100
n = np.arange(n_samples)
x_n = np.sin(0.2 * np.pi * n)
y_n = np.zeros(n_samples)

for i in range(1, n_samples):
    y_n[i] = 0.7 * y_n[i-1] + x_n[i]

plt.figure(figsize=(10, 6))
plt.plot(n, x_n, label='Input signal x[n],', color='blue')
plt.plot(n, y_n, label='Output signal y[n]', color='green')
plt.title('Linear-Time-Invariant (LTI) System Response')
plt.xlabel('n (samples)')
plt.ylabel('Signal Value')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```

## Problem 2

```python
n_samples = 100
n = np.arange(n_samples)
x_n = np.sin(0.2 * np.pi * n)
y_n = x_n ** 2

plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.scatter(n, x_n, color='blue', label = 'Input Signal x[n]', alpha = 0.6)
plt.title('Scatter Plot of Input Signal')
plt.xlabel('Sample Number (n)')
plt.ylabel('x[n]')
plt.grid()
plt.legend()

plt.subplot(1,2,2)
plt.bar(n, y_n, color='orange', label = 'Output Signal y[n]', alpha = 0.6)
plt.title('Bar Plot of Output Signal')
plt.xlabel('Sample Number (n)')
plt.ylabel('y[n]')
plt.grid()
plt.legend()

plt.figure(figsize = (10,5))
plt.plot(n, x_n, color='blue', label = 'Input Signal x[n]')
plt.plot(n, y_n, color='orange', label = 'Output Signal y[n]')
plt.title('Non-Linear System Response')
plt.xlabel('Sample Number (n)')
plt.ylabel('Signal Value')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```

## Problem 3

```python
n = np.arange(100)
t = np.linspace(0, 2*np.pi, 100)
x_clean = np.sin(0.3 * np.pi * n)
noise = np.random.normal(0, 0.1, 100)
x_noisy = x_clean + noise

y = np.zeros(100)
for i in range(1, 100):
    y[i] = 0.5 * y[i-1] + 0.5 * x_noisy[i]


plt.figure(figsize=(10, 6))
plt.plot(n, x_clean, label='Clean Signal', color='blue')
plt.title('Clean Signal')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()


plt.figure(figsize=(10, 6))
plt.plot(n, x_noisy, label='Noisy Input Signal', color='red')
plt.title('Noisy Input Signal')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()


plt.figure(figsize=(10, 6))
plt.plot(n, y, label='System Output', color='green')
plt.title('System Output to Noisy Input Signal')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```
