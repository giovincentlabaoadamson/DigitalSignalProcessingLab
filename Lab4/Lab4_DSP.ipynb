# Laboratory Experiment 3

### Signal Generation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(t):
    y = np.exp(-t/2)
    return y
```

### Signal as Functions

```python
X = np.linspace(0,10,1000)
plt.plot(X, f(X))
plt.show()
```

### Signal as Vectors

```python
T = np.arange(0,8)
Y = np.array([0, 1e-3, 1e-2, 1e-1, 1, 1e1, 1e2, 1e3])
plt.plot(T,Y)
plt.show()
```

### Sampling

```python
f = np.sin
N_low = np.linspace(0, 2*np.pi, 5)

plt.stem(N_low, f(N_low))
plt.plot(N_low, f(N_low), color='blue')

plt.show()
```

```python
N_high = np.linspace(0, 2*np.pi, int(1e6))

plt.stem(N_high, f(N_high))
plt.plot(N_high, f(N_high), color='blue')

plt.show()
```

# Laboratory Experiment 4

## Aperiodic Signal

### Impulse Signal

```python
N = np.linspace (-10,10,21)

def impulse(N):
    return np.where(N == 7, 8, 1)
plt.stem(N, impulse(N))
plt.grid()
plt.show()
```

### Ramp Signal

```python
def ramp(t,m,b):
    return m*t+b
plt.plot(N, ramp(N, 1,7), label='m=1, b=7')
plt.plot(N, ramp(N, 5,3), label='m=5, b=3')
plt.plot(N, ramp(N, -2,8), label='m=-2, b=8')
plt.plot(N, ramp(N, -0.5,0), label='m=0.5, b=0')

plt.axhline(color='k')
plt.axvline(color='k') 

plt.legend()
plt.grid()
plt.show()
```

### Exponential Signal

```python
N = np.linspace(-3, 3, 21)
plt.stem(N, np.exp(N), linefmt='orange', label=r'$e^n$')
plt.stem(N, np.exp(-N), linefmt='blue', label=r'$e^{n}$')
plt.stem(N, -np.exp(N), linefmt='orangered', label=r'$-e^{n}$')
plt.stem(N, -np.exp(-N), linefmt='dodgerblue', label=r'$-e^{-n}$')

plt.legend()
plt.grid()
plt.show()
```

## Periodic Signal

### Unit Signal

```python
N = np.linspace(-10, 10, 21)

def step(N):
    return np.where(N >= 2.5, 2, 3)

plt.stem(N, step(N))

plt.grid()
plt.show()
```

### Sinusodial Signal

#### Amplitude Shift

```python
N = np.linspace(-1*np.pi,1*np.pi , 50)

plt.figure(figsize=(10,5))
plt.title('Amplitude Shift')
plt.plot(N, np.sin(N), label=r'$\sin \theta$')
plt.plot(N, -5*np.sin(N), label=r'$-5\sin \theta$')

plt.legend()
plt.grid()
plt.show()
```

#### Frequency Shift

```python
plt.figure(figsize=(10,5))
plt.title('Frequency Shift')
plt.plot(N, np.sin(N), label=r'$\sin \theta$', color = 'blue')
plt.plot(N, np.sin(-5*N), label=r'$\sin -5\theta$', color ='yellow')
plt.plot(N, np.sin(8*N), label=r'$\sin 8\theta$', color = 'red')

plt.legend()
plt.grid()
plt.show()
```

#### Phase Shift

```python
plt.figure(figsize=(10,5))
plt.title('Phase Shift')

plt.plot(N, np.sin(N), label=r'$\sin \theta$')
plt.plot(N, np.sin(N+np.pi/-8), label=r'$\sin \theta + \frac{\pi}{-8}$')
plt.plot(N, np.sin(N+np.pi/7), label=r'$\sin \theta + \frac{\pi}{7}$')

plt.legend()
plt.grid()
plt.show()
```

### Euler's Formula and Identity

```python
def euler(theta):
    return np.exp(1j*theta)
    
N = np.linspace(-5*np.pi,5*np.pi , 100)

plt.figure(figsize=(10,5))

plt.plot(N, euler(N).real, label='real')
plt.plot(N, euler(N).imag, label='imaginary')

plt.legend()
plt.grid()
plt.show()
```
