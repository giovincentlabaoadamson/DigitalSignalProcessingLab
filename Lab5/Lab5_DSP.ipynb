```python
import numpy as np
import matplotlib.pyplot as plt
```

# Laboratory Activity 5

## Problem 1

```python
t = np.linspace (0,10,100)

x_t = np.sin(2*np.pi*t)
y_t = 2 * x_t

t_shift = 2 - x_t
x_t_shift = 2 * t_shift
y_t_shift = 2 * x_t_shift

plt.subplot (2,1,1)
plt.plot (t, x_t, label ='Input Signal x(t)', color = 'blue')
plt.plot (t, y_t, label ='Output Signal y(t)', color = 'red')
plt.title ('Original Input and Output Signals')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()

plt.subplot (2,1,2)
plt.plot (t, x_t_shift, label ='Shifted Input Signal x(t)', color = 'green')
plt.plot (t, y_t_shift, label ='Shifted Output Signal y(t)', color = 'orange')
plt.title ('Shifted Input and Output Signals')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()

```

## Problem 2

```python
t = np.linspace(0, 10, 1000)
               
x_t = np.sin(t)
y_t = t * x_t

t_shift = 2 - t
x_shift = np.sin(t_shift)
y_shift = t_shift * x_shift

plt.subplot (2,1,1)
plt.plot(t, x_t, label="Input Signal x(t)")
plt.plot(t, y_t, label="Output Signal y(t)")
plt.title("Total Cost of Electricity y(t)")
plt.xlabel("Time (hours)")
plt.ylabel("Cost")
plt.legend()
plt.grid(True)
plt.show()

plt.subplot(2,1,2)
plt.plot(t, x_shift, label="Shifted Input Signal x(t)")
plt.plot(t, y_shift, label="Shifted Output Signal y(t)")
plt.title("Shifted Total Cost of Electricity y(t - tau)")
plt.xlabel("Time (hours)")
plt.ylabel("Cost")
plt.legend()
plt.grid(True)
plt.show()
```

## Problem 3

```python
t = np.linspace(0, 10, 1000)
x_t = np.sin(t)
dt = t[1] - t[0]
y_t = np.gradient(x_t, dt)

tau = 2
x_shifted = np.sin(t - tau)
y_shifted = np.gradient(x_shifted, dt)


plt.subplot(2, 1, 1)
plt.plot(t, x_t, label="Position x(t) = sin(t)")
plt.plot(t, y_t, label="Velocity y(t) = d/dt[x(t)]")
plt.title('Position and Velocity of the Car')
plt.xlabel('Time [s]')
plt.ylabel('Position / Velocity')
plt.legend()
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(t, x_shifted, label="Shifted Position x(t-2)")
plt.plot(t, y_shifted, label="Shifted Velocity y(t-2) = d/dt[x(t-2)]")
plt.title('Shifted Position and Velocity of the Car (2s delay)')
plt.xlabel('Time [s]')
plt.ylabel('Position / Velocity')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
