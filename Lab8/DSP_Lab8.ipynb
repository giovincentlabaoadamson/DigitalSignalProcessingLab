## Basic Signal Analysis with the CFT 

```python
import numpy as np 

import matplotlib.pyplot as plt 

 

t = np.linspace(-1, 1, 500) 

f0 = 5  # Frequency of the signal 

 

signal = np.sin(2 * np.pi * f0 * t) 

 

signal_fft = np.fft.fft(signal) 

frequencies = np.fft.fftfreq(len(signal), d=(t[1] - t[0])) 

 

plt.figure(figsize=(12, 5)) 

 

plt.subplot(1, 2, 1) 

plt.plot(t, signal) 

plt.title("Time Domain Signal") 

plt.xlabel("Time (s)") 

plt.ylabel("Amplitude") 

 

plt.subplot(1, 2, 2) 

plt.plot(frequencies, np.abs(signal_fft)) 

plt.title("Frequency Domain (Magnitude)") 

plt.xlabel("Frequency (Hz)") 

plt.ylabel("Magnitude") 

 

plt.tight_layout() 

plt.show() 
```

## Signal Composition and Frequency Analysis 

```python
f1, f2 = 5, 10 

composite_signal = np.sin(2 * np.pi * f1 * t) + np.sin(2 * np.pi * f2 * t) 

 

composite_fft = np.fft.fft(composite_signal) 

 

plt.figure(figsize=(12, 5)) 

 

plt.subplot(1, 2, 1) 

plt.plot(t, composite_signal) 

plt.title("Time Domain Composite Signal") 

plt.xlabel("Time (s)") 

plt.ylabel("Amplitude") 

 

# Frequency Domain 

plt.subplot(1, 2, 2) 

plt.plot(frequencies, np.abs(composite_fft)) 

plt.title("Frequency Domain of Composite Signal") 

plt.xlabel("Frequency (Hz)") 

plt.ylabel("Magnitude") 

 

plt.tight_layout() 

plt.show() 
```

## Effect of Time Shifting on the CFT 

```python
t_shift = 0.2 

 

shifted_signal = np.cos(2 * np.pi * f0 * (t - t_shift)) 

 

shifted_fft = np.fft.fft(shifted_signal) 

 

plt.figure(figsize=(12, 5)) 

 

plt.subplot(1, 2, 1) 

plt.plot(t, shifted_signal) 

plt.title("Time Domain: Shifted Signal") 

plt.xlabel("Time (s)") 

plt.ylabel("Amplitude") 

 

plt.subplot(1, 2, 2) 

plt.plot(frequencies, np.abs(shifted_fft)) 

plt.title("Frequency Domain: Shifted Signal") 

plt.xlabel("Frequency (Hz)") 

plt.ylabel("Magnitude") 

 

plt.tight_layout() 

plt.show() 
```

## Generating Synthetic Wave 

```python
import numpy as np 

from scipy.io.wavfile import write 

 

# Define parameters 

sample_rate = 44100  # samples per second 

duration = 2  # seconds 

frequency = 440  # A4 note in Hz (440 Hz) 

 

# Generate time values 

t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False) 

# Generate sine wave signal 

signal = 0.5 * np.sin(2 * np.pi * frequency * t) 

 

# Save the signal as a .wav file 

write("sample.wav", sample_rate, signal.astype(np.float32)) 
```
